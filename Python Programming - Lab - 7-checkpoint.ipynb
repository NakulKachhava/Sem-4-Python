### 02) WAP to find Maximum and Minimum K elements in a given tuple.
**Sample Input:**
```
T = (7, 2, 9, 4, 1, 5)
K = 2

```
**Sample Output:**
```
Minimum 2 elements: (1, 2)
Maximum 2 elements: (9, 7)
```

````python
len1 = int(input("Enter Length: "))

t1 = tuple((int(i) for i in input().split()))

k = int(input("Enter K elements: "))

t1 = sorted(t1)

print(f"Minimum {k} elements: {tuple(t1[0:k])}")
print(f"Maximum {k} elements: {tuple(t1[-1:-k-1:-1])}")
````

### 03) WAP to find tuples which have all elements divisible by K from a list of tuples.
**Sample Input:**
```
tuple_list = [(2, 4, 6), (3, 6, 9), (4, 8, 12), (5, 10, 15)]
K = 2

```
**Sample Output:**
```
Tuples divisible by 2 : [(2, 4, 6), (4, 8, 12)]
```

````python
len2 = int(input("Enter Length: "))
li = []

for i in range(len2):
    t2 = tuple((int(i) for i in input().split()))
    li.append(t2)

k1 = int(input("Enter K elements: "))

l1 = []

for i in li:
    for j in i:
        if(j%k1 != 0):
            break
    else:
        l1.append(i)
        
print(f"Tuples divisible by {k1} : {l1}")
````

### 05) WAP to find tuples with all positive elements from the given list of tuples.

```

```
