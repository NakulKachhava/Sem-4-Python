### 03) Write a function that returns 1 if the given number is Prime or 0 otherwise.

```python
def isPrime(n):
    for i in range(2,n):
        if n%i == 0:
            return 0
    else:
        return 1
    
n = int(input("Enter a number: "))

ans = "prime" if isPrime(n) == 1 else "not prime"

print(f"{n} is {ans}")
```

### 04) Write a function that returns the list of Prime numbers between given two numbers.

```python
def primeList(a,b):
    li = []
    for i in range(a,b+1):
        for j in range(2,i):
            if i%j == 0:
                break
        else:
            li.append(i)
    return li

a = int(input("Enter lower bound: "))

b = int(input("Enter upper bound: "))

print(f"List of prime numbers: {primeList(a,b)}")
```

### 11) Write a function to invert a given Dictionary.
#### hint: keys to values & values to keys
#### Before : {'a': 10, 'b':20, 'c':30, 'd':40}
#### After : {10:'a', 20:'b', 30:'c', 40:'d'}

```python
def invert(d):
    d1 = {}
    for k,v in d.items():
        d1[v] = k
    return d1

k2  = list(input("Enter keys for d1 seperated with space: ").split(" "))
v2 = list(map(int,input("Enter values for d1 seperated with space: ").split(" ")))

d2 = dict(zip(k2,v2))

print(f"Inverted dictionary: {invert(d2)}")
```

### 12) Write a function that returns the number of uppercase and lowercase letters in the given string. 
#### example : Input : s1 = AbcDEfgh ,Ouptput : no_upper = 3, no_lower = 5

```python
def noOfUppLow(s):
    upp = 0
    low = 0
    for i in s:
        if i.isupper():
            upp += 1
        else:
            low += 1
    return upp, low

s1 = input("Enter a string: ")

upp, low  = noOfUppLow(s1)

print(f"no_upper = {upp}, no_lower = {low}")
```

### 14) For the given list of names of students, extract the names having more that 7 characters. Use filter().

```python
def moreThanSeven(i):
    if len(i) > 7:
        return True
    else:
        return False

li5 = list(input("Enter names of students seperated with space: ").split(" "))

print(f"Names of students having more than 7 characters: {list(filter(moreThanSeven,li5))}")
```
