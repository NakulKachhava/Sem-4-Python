### 05) WAP to split the List into two parts and append the first part to the end.

```python
li = list(map(int,input("Enter list space seperated: ").split(" ")))

midIndex = len(li) // 2

fp = li[:midIndex]
sp = li[midIndex:]

l1 = sp + fp

print(l1)
```

### 09) WAP to count unique items in a list.

```python
li = list(map(int,input("Enter list space seperated: ").split(" ")))

unique = []
count = 0

for i in li:
    if i not in unique:
        count += 1
        unique.append(i)

print(f"Unique Items: {count}")
```
