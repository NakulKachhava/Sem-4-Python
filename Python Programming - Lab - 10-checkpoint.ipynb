### 08) WAP to create an UDF named frequency to count occurances of the specific word in a given text file.

```python
def frequency(l,w):
    return l.count(w)

fp = open("helloworld.txt","r")
noOfLines = fp.readlines()
li1 = []
for i in noOfLines:
    for j in i.split(" "):
        li1.append(j)
        
w = input("Enter a specific word: ")

print(f"Occurances of {w} in the file {frequency(li1,w)}")
```
