### 01) WAP to find their union, intersection, and difference of the given two sets.
**Sample Input:**
```
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

```
**Sample Output:**
```
Union: {1, 2, 3, 4, 5, 6}
Intersection: {3, 4}
Difference (A-B): {1, 2}

```

### 05) WAP to Merge two dictionaries. If a key exists in both dictionaries, add their values.

**Sample Input:**
```
Dict1 = {1: 10, 2: 20, 3: 30}
Dict2 = {2: 40, 3: 50, 4: 60}

```
**Sample Output:**
```
{1: 10, 2: 60, 3: 80, 4: 60}
```

````python
k1 = list(map(int,input("Enter keys for d1 seperated with space: ").split(" ")))
v1  = list(map(int,input("Enter values for d1 seperated with space: ").split(" ")))

d1 = dict(zip(k1,v1))

k2 = list(map(int,input("Enter keys for d2 seperated with space: ").split(" ")))
v2 = list(map(int,input("Enter values for d2 seperated with space: ").split(" ")))

d2 = dict(zip(k2,v2))

if len(d1) > len(d2):
    for k2,v2 in d2.items():
        if k2 in d1.keys():
            d1[k2] = d1[k2] + d2[k2]
        else:
            d1[k2] = d2[k2]
    print(d1)
else:
    for k1,v1 in d1.items():
        if k1 in d2.keys():
            d2[k1] = d2[k1] + d1[k1]
        else:
            d2[k1] = d1[k1]
    print(d2)
````

### 06) WAP to Sort a dictionary in ascending order of its values.
Hint: in sorted(), use udf for key parameter


**Sample Input:**
```
{'Math': 70, 'Physics': 85, 'Chemistry': 60}

```
**Sample Output:**
```
{'Chemistry': 60, 'Math': 70, 'Physics': 85}
```

````python
k3 = list(input("Enter keys for d3 seperated with space: ").split(" "))
v3  = list(map(int,input("Enter values for d3 seperated with space: ").split(" ")))

d3 = dict(zip(k3,v3))

def val(item):
    return item[1]

print(dict(sorted(d3.items(), key=val)))
````

### 07) WAP to find the keys having maximum and minimum values in a dictionary.

**Sample Input:**
```
{'A': 50, 'B': 90, 'C': 30}
```
**Sample Output:**
```
Max Value Key: B
Min Value Key: C

```

````python
k4 = list(input("Enter keys for d4 seperated with space: ").split(" "))
v4  = list(map(int,input("Enter values for d4 seperated with space: ").split(" ")))

d4 = dict(zip(k4,v4))

maximum = max(d4.values())
minimum = min(d4.values())

for key,val in d4.items():
    if val == maximum:
        print(f"Max Value Key: {key}")
    if val == minimum:
        print(f"Min Value Key: {key}")
````
